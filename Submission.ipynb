## Cleaning Data

Proses mengidentifikasi dan mengoreksi atau membuang data yang salah dari dataset yang digunakan

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
  hour_df[column] = pd.to_datetime(day_df[column])
hour_df.info()
```

Melakukan koreksi pada tipe data sehingga tipe data yang akan digunakan menjadi sesuai

```python
datetime_columns = ["dteday"]
for column in datetime_columns:
  day_df[column] = pd.to_datetime(day_df[column])
day_df.info()
```

## Exploratory Data Analysis ( EDA )

```python
day_df.describe(include="all")
```

```python
total_per_musim = day_df.groupby('season')['cnt'].sum().sort_values(ascending=False)

# Menampilkan total penyewa sepeda per musim
print("Total Penyewa Sepeda per Musim:")
print(total_per_musim)
```

```python
total_per_hari_seluruh_musim = day_df.groupby([ 'weekday'])['cnt'].mean()

# Menampilkan total penyewa sepeda pada musim 1 per hari dengan menampilkan hari dalam seminggu
print("Rata Rata Penyewa Sepeda pada Seluruh Musim Pada Hari Tertentu:")
print(total_per_hari_seluruh_musim)
```

```python
total_rentals_per_season = day_df.groupby('season')[['casual', 'registered']].sum()
print(total_rentals_per_season)
```

```python
matriks_korelasi = day_df[['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(matriks_korelasi, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Matriks Korelasi antara Variabel Numerik')
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1

Pada hari apa penyewaan sepeda menempati posisi tinggi

```python
plt.figure(figsize=(10, 6))
total_per_hari_seluruh_musim.plot(kind='bar', color='skyblue')
plt.title('Rata rata Hari Penyewaan Sepeda pada Seluruh Musim')
plt.xlabel('Weekday')
plt.ylabel('Total Penyewa Sepeda')
plt.xticks(rotation=0)  # Rotasi label x agar lebih mudah dibaca
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### Pertanyaan 2

Pada musim apa penyewaan sepeda tertinggi

```python
plt.figure(figsize=(10, 6))
total_per_musim.plot(kind='bar', color='skyblue')
plt.title('Total Penyewa Sepeda Permusim')
plt.xlabel('Season')
plt.ylabel('Total Penyewa Sepeda')
plt.xticks(rotation=0)  # Rotasi label x agar lebih mudah dibaca
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
average_rentals_per_weather_season = day_df.groupby(['weathersit', 'season'])['cnt'].mean()

# Mapping kode cuaca menjadi teks untuk kejelasan pada plot
weather_mapping = {
    1: 'Cerah',
    2: 'Kabut',
    3: 'Hujan Ringan',
    4: 'Hujan Lebat'
}

# Mapping kode musim menjadi teks untuk kejelasan pada plot
season_mapping = {
    1: 'Spring',
    2: 'Summer',
    3: 'Fall',
    4: 'Winter'
}

# Menyiapkan data untuk plot
plot_data = average_rentals_per_weather_season.unstack().rename(columns=season_mapping)

# Menampilkan bar chart
plt.figure(figsize=(10, 6))
plot_data.rename(index=weather_mapping).plot(kind='bar', color=['skyblue', 'orange', 'green', 'purple'])
plt.title('Rata-rata Penyewaan Sepeda pada Kondisi Cuaca untuk Setiap Musim')
plt.xlabel('Kondisi Cuaca')
plt.ylabel('Rata-rata Penyewaan Sepeda')
plt.xticks(rotation=45)  # Rotasi label x agar lebih mudah dibaca
plt.legend(title='Season')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

## Conclusion 

- Pertanyaan 1: Berdasarkan rata rata, penyewaan sepeda tertinggi pada hari 5 (Jumat)
- Pertanyaan 2: Penyewaan sepeda tertinggi terjadi pada musim 3 (Fall)
